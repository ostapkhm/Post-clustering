```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import norm

mpl.rcParams['figure.dpi'] = 300
plt.style.use('seaborn-v0_8-muted')

# Define the parameters for the normal distributions
mean1 = 0
mean2 = 2
variance1 = 1
variance2 = 1

# Generate points for the x-axis
x = np.linspace(-5, 6, 1000)

# Calculate the probability density functions (PDFs) of the two normal distributions
pdf1 = norm.pdf(x, mean1, np.sqrt(variance1))
pdf2 = norm.pdf(x, mean2, np.sqrt(variance2))

# Calculate the mixture components
mixture_component1 = 0.5 * pdf1
mixture_component2 = 0.5 * pdf2

# Calculate the sum of the mixture components
mixture_sum = mixture_component1 + mixture_component2

# Plot the individual mixture components and their sum
plt.plot(x, mixture_component1, label=f'$0.5 \: N({mean1}, {variance1})$')
plt.plot(x, mixture_component2, label=f'$0.5 \: N({mean2}, {variance2})$')
plt.plot(x, mixture_sum, label='Суміш')


plt.title('Двокомпонента одновимірна гаусова суміш')
plt.xlabel('x')
plt.ylabel('Щільність')
plt.legend()
plt.grid(True)
```
